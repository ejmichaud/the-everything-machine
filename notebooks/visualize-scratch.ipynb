```python
import json

import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("../tmp/16/config.json", 'r') as f:
    config = json.load(f)
with open("../tmp/16/info.json", 'r') as f:
    info = json.load(f)
```

```python
Ss = info['Ss']


plt.plot(info['log_steps'], info['losses'], color='red', label="total loss")

colors = plt.cm.viridis(np.linspace(0.1, 0.9, len(Ss)))
for i in range(len(Ss)):
    plt.plot(info['log_steps'], info['losses_subtasks'][str(i)], color=colors[i], alpha=0.3, zorder=-10)

# plt.xlim(0, 1000)
plt.ylabel("Cross-entropy loss")
plt.xlabel("Optimization Steps")
# plt.xscale('log')
plt.legend()
```
